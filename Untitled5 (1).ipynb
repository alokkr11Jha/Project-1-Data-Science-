```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame with 'pre molt size' and 'post molt size' columns

# Step 1: Clean Data - Drop rows with NaN values if any
df_cleaned = df.dropna().loc[~np.isinf(df).any(axis=1)]

# Step 2: Calculate means and standard deviations
mean_post = np.mean(df_cleaned['post molt size'])
mean_pre = np.mean(df_cleaned['pre molt size'])
std_post = np.std(df_cleaned['post molt size'], ddof=1)  # ddof=1 for sample standard deviation
std_pre = np.std(df_cleaned['pre molt size'], ddof=1)

# Step 3: Calculate pooled standard deviation
pooled_std = np.sqrt((std_post**2 + std_pre**2) / 2)

# Step 4: Calculate Cohen's d
cohen_d = (mean_post - mean_pre) / pooled_std
print(f"Cohen's d: {cohen_d}")
```
